```python
%matplotlib inline
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
Line =  np.poly1d( [ 1,  0])
### plotting the polynomial Function (red line)
ls = np.linspace(0, 1000, 1000)


### Labels
plt.xlabel(r'PV $\longrightarrow$')
plt.ylabel(r'MV $\longrightarrow$')
#ax.set_ylabel(r'$\Delta \Theta / \omega \longrightarrow$')
plt.title("Positive Linear Correlation")

### the Line is drawn
plt.plot(ls, Line(ls), c='r')

### printing the graph
plt.show()
```

```python
Line =  np.poly1d( [ -1,  0])
### plotting the polynomial Function (red line)
ls = np.linspace(0, 1000, 1000)


### Labels
plt.xlabel(r'PV $\longrightarrow$')
plt.ylabel(r'MV $\longrightarrow$')
plt.title("Negative Linear Correlation")

### the Line is drawn
plt.plot(ls, Line(ls), c='r')

### printing the graph
plt.show()
```

```python
Line =  np.poly1d( [ 10, 0,  0])
### plotting the polynomial Function (red line)
ls = np.linspace(0, 10, 100)


### Labels
plt.xlabel(r'PV $\longrightarrow$')
plt.ylabel(r'MV $\longrightarrow$')
plt.title("Quadratic Correlation")

### the Line is drawn
plt.plot(ls, Line(ls), c='r')

### printing the graph
plt.show()
```

```python
Line =  np.poly1d( [ 10, -100, 1, -100, 100 ])
### plotting the polynomial Function (red line)
ls = np.linspace(0, 10, 100)

### Labels
plt.xlabel(r'PV $\longrightarrow$')
plt.ylabel(r'MV $\longrightarrow$')
plt.title("Polynomial Correlation")

### the Line is drawn
plt.plot(ls, Line(ls), c='r')

### printing the graph
plt.show()
```
